# Data Tranformation

```python
df.head()
```

```python
def remove_space(text):
    return str(text).replace(" ", "")
```

```python
def list_convert(text):
    return str(text).split(",")
```

```python
df['title'] = df['title'].apply(remove_space)
df['genres'] = df['genres'].apply(remove_space)
df['original_title'] = df['original_title'].apply(remove_space)
df['genres'] = df['genres'].apply(list_convert)
```

```python
import re

def clean_text_to_five(text):
    cleaned = re.sub(r'[^A-Za-z0-9\s]', '', str(text))
    words = cleaned.split()
    return words
```

```python
df['genres'] = df['genres'].apply(clean_text_to_five)
```

```python
def make_tags(row):
    genres = row['genres'] if isinstance(row['genres'], list) else [row['genres']]
    title = [str(row['title'])] if pd.notnull(row['title']) else []
    imdb_id = [str(row['imdb_id'])] if pd.notnull(row['imdb_id']) else []
    year = [str(int(row['year']))] if pd.notnull(row['year']) else []
    return title + genres + year + imdb_id 

df['tags'] = df.apply(make_tags, axis=1)
```

```python
df2 = df[['imdb_id', 'title', 'tags']]
df2.head(10)
```

```python
df2['tags'] = df2['tags'].apply(lambda x:" ".join(x))
```
